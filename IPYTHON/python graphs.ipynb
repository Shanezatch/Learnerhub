```python
import pandas as pan
import numpy as ny
import matplotlib.pyplot as matpa
raw=pan.read_csv("1. Regression - Module - (Housing Prices).csv")
matpa.plot(raw["Sale Price"],color="green")
matpa.xlabel("Sale price")
matpa.ylabel("No of House")
matpa.title('Sale price of house using line graph')
#matpa.show()
```

```python
matpa.scatter(x=raw["Condition of the House"],y=raw["Sale Price"],color="blue")
matpa.xlabel("Condition of the house")
matpa.ylabel("Price of the house")

matpa.title("price on the condition of the house ")
matpa.show()
```

```python
matpa.hist(raw["Age of House (in Years)"],bins=10,linewidth=50)
matpa.xlabel("Age of house in years")
matpa.ylabel("No of records")
matpa.title("Age of house in histogram graph")
matpa.show()
```

```python
matpa.boxplot(raw["Age of House (in Years)"])
matpa.xlabel("Age of house in years")
matpa.ylabel("No of records")
matpa.title("Age of house in boxplot graph")
matpa.show()
```

```python
raw.groupby('Condition of the House')['ID'].count() 
values=(30,1701,14031,5679,702)
labels=("'bad","Excellent","Fair","Good","Okay")
matpa.pie(values,labels = labels)
```
